Requires `mlagents_envs` to be installed.

The `mlagents_envs` package in this directory is a copy of the [official GitHub repository](https://github.com/Unity-Technologies/ml-agents/tree/develop/ml-agents-envs/mlagents_envs), **but with a recompiled protobuf**.

The official version depends on `protobuf>=3.6,<3.21`, while this copy has been recompiled with `protobuf==6.33.2`.

If you prefer to use the official protobuf version, please remove the `mlagents_envs` directory here and use the official package instead.

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from unity_wrapper import UnityWrapper

np.set_printoptions(threshold=140)
%matplotlib inline
```

```python
env = UnityWrapper(train_mode=True, base_port=5004)
```

```python
ma_obs_names, ma_obs_shapes, ma_d_action_sizes, ma_c_action_size = env.init()
print(ma_obs_names)
print(ma_obs_shapes)
print(ma_d_action_sizes)
print(ma_c_action_size)
```

```python
ma_agent_ids, ma_obs_list = env.reset()

for step in range(10):
    print('=' * 10, 'step', step, '=' * 10)

    ma_d_action = {}
    ma_c_action = {}

    for n, obs_list in ma_obs_list.items():
        print('[ma_name] :', n)
        obs_shapes = ma_obs_shapes[n]
        d_action_sizes = ma_d_action_sizes[n]
        c_action_size = ma_c_action_size[n]

        agents_size = obs_list[0].shape[0]
        if agents_size == 0:
            continue

        for obs_i, obs in enumerate(obs_list):
            print('--obs_name-- :', ma_obs_names[n][obs_i], ma_obs_shapes[n][obs_i])
            if len(obs.shape) == 4:
                fig, axes = plt.subplots(ncols=agents_size, squeeze=False, figsize=(1 * agents_size, 1))
                for i, image in enumerate(obs):
                    if image.shape[0] == 1:
                        axes[0][i].imshow(image.transpose(1, 2, 0), cmap='gray', vmin=0, vmax=1)
                    else:
                        axes[0][i].imshow(image.transpose(1, 2, 0))
                    axes[0][i].axis('off')

                    # np.save(f'obs/{step}-{n.replace("?", "_")}-{ma_obs_names[n][obs_i]}-{i}.npy', image)
                    # if image.shape[0] == 1:
                    #     pass
                    # else:
                    #     pil_image = Image.fromarray((image.transpose(1, 2, 0) * 255.).astype(np.uint8), 'RGB')
                    #     pil_image.save(f'obs/{step}-{n.replace("?", "_")}-{ma_obs_names[n][obs_i]}-{i}.png')

                # plt.savefig(f'obs/{step}-{n.replace("?", "_")}-{ma_obs_names[n][obs_i]}.png', bbox_inches='tight', pad_inches=0.0)
                plt.show()
            else:
                print(obs)

        d_action, c_action = None, None
        if ma_d_action_sizes[n]:
            d_action_list = [np.random.randint(0, d_action_size, size=agents_size)
                             for d_action_size in ma_d_action_sizes[n]]
            d_action_list = [np.eye(d_action_size, dtype=np.int32)[d_action]
                             for d_action, d_action_size in zip(d_action_list, ma_d_action_sizes[n])]
            d_action = np.concatenate(d_action_list, axis=-1)

        if ma_c_action_size[n]:
            c_action = np.random.randn(agents_size, ma_c_action_size[n])

        ma_d_action[n] = d_action
        ma_c_action[n] = c_action

    (decision_step,
     terminal_step,
     all_envs_done) = env.step(ma_d_action,
                               ma_c_action)

    ma_obs_list = decision_step.ma_obs_list
```

```python
env.close()
```
